```python
#Que#06
#Write a program which takes 5 inputs from user for different subject’s marks, total it and generate mark sheet using grades ?
eng = int(input("Enter your English marks : "));
isl = int(input("Enter your Islamiat marks : "));
maths = int(input("Enter your Maths marks : "));
urdu = int(input("Enter your Urdu marks : "));
sci = int(input("Enter your Science marks : "));
ObtMarks = eng + isl + maths + urdu + sci;
TtlMarks = 425;
Per = ObtMarks/TtlMarks * 100;
if Per < 100 and Per >= 80:
    print("Grade A+");
elif Per < 80 and Per >= 70:
    print("Grade A");
elif Per < 70 and Per >= 60:
    print("Grade B");
elif Per < 60 and Per >= 50:
    print("Grade C");
elif Per < 50 and Per >= 40:
    print("Grade D");
elif Per < 40 and Per >= 33:
    print("Grade E");
elif Per < 33 and Per >= 0:
    print("Fail");
else:
    print("The percentage is inappropriate");
```
